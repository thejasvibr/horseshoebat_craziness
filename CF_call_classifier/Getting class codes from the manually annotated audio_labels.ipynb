I have a manually annotated set of audio data from the experimentally collected data, and would now like to test the prediction of the networks with it. However I still ned to clean up the data because i labelled it wrong. 

```python
import numpy as np 
import pandas as pd
```

```python
audio_lab = pd.read_csv('audio_labels.csv')
```

```python
audio_lab
```

```python
#replace the 0's with O's because they won't get messed up by spreadsheets
for column in ['Ferrum_call','BlEuMi_call','Myotis_call']:
    rows_w_zeros = audio_lab[column]==0
    audio_lab[column]= audio_lab[column].astype(str)
    audio_lab[column][rows_w_zeros] = 'O'
```

```python
audio_lab.head()
```

```python
single_audio_lab = audio_lab[audio_lab['groupsize_label']=='single']
single_audio_lab.head()
```

```python
# give single entries the class names:
for i, row in single_audio_lab.iterrows():
        single_audio_lab.loc[i,'class_name'] = row['Ferrum_call']+row['BlEuMi_call']+row['Myotis_call']
```

```python
single_audio_lab.head()
```

```python
# get all entries with multiple bat situations
multi_audio_lab = audio_lab[audio_lab['groupsize_label']=='multi']
multi_audio_lab.head()
```

```python
# manualy annotate the situations where only a single CF bat species is present:
for i, row in multi_audio_lab.iterrows():
    if sum([row['Ferrum_call']=='1',row['BlEuMi_call']=='1']) == 1:
            if row['Ferrum_call'] == '1' :
                multi_audio_lab.loc[i,'class_name'] = 'mO'+row['Myotis_call']
            elif row['BlEuMi_call'] == '1':
                multi_audio_lab.loc[i,'class_name'] = 'Om'+row['Myotis_call']
            
            
```

```python
multi_audio_lab
```

```python
none_labels = audio_lab[audio_lab['groupsize_label']=='none']
```

```python
none_labels
```

```python
for i, row in none_labels.iterrows():
    none_labels.loc[i,'class_name'] = 'OO' + row['Myotis_call']
```

```python
# combine the single and multi bat species dataframes :
cleaned_labels = pd.concat((single_audio_lab,multi_audio_lab,none_labels))
```

```python
cleaned_labels
```

```python
cleaned_labels.to_csv('audio_labels_cleaned.csv')
```
